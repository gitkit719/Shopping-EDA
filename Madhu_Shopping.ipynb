#### Cleaning all null values by using Dropna

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ["CustomerID","Gender","Location","Tenure_Months" ,"Transaction_ID", "Transaction_Date", "Product_SKU", "Product_Description"]) 
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Coupon_Code", "Discount_pct"])
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

#### Seperating Year from Date column
*Month column is already in the dataset so no need to extract Month                                                              
*DayofMonth is not required for this analysis

```python
df["Year"] = pd.to_datetime(df["Date"]).dt.year
df.head()
```

#### Droping the unnecessary columns

```python
df = df.drop(["Transaction_Date", "Date"], axis=1)
df
```

#### Comment

#### Product Trends:					
Apparel dominates as the most popular category, while Waze exhibits the least popularity.					
Both male and female categories contribute equally to overall product demand.					
#### Month-wise Insights:					
October sees peak demand, particularly for "Nest-USA."					
January, notably in New Jersey, experiences lower demand, suggesting potential seasonality.					
#### Coupon Engagement:					
"Clicked" status reflects the highest consumer interaction.					
Offline spending consistently surpasses online spending across all coupon statuses.					
#### Geographical Dynamics:					
Chicago stands out as a high-demand location, with Washington DC experiencing lower demand.					
Delivery charges correlate with demand, with Chicago having the highest charges for the "Male" category.					
#### Product Quantity Analysis:					
Certain products show uniform consumption between male and female categories.					
"Office" emerges as the most popular product among females.
